# All AL TCs

```python
hurdata = data[data['Basin'].str.contains('NA',na=False)]
```

```python
hurdata
```

```python
hur=hurdata.groupby('Serial_Num')#.apply(tuple)
```

```python
ids = hur.groups.keys()
```

```python
len(ids)
```

```python
mapa = folium.Map([24.,-60.],zoom_start=2)
```

```python
for i in range(len(ids)):
    lat = hur.get_group(ids[i])['Latitude'].apply(pd.to_numeric)
    y = np.ma.masked_array(lat, lat==-999.).compressed()
    lon = hur.get_group(ids[i])['Longitude'].apply(pd.to_numeric)
    x = np.ma.masked_array(lon,lon==-999.).compressed()
    points = [list(i) for i in zip(y, x)]
    
    folium.PolyLine(points, color="red", weight=2.5, opacity=1).add_to(mapa)
```
